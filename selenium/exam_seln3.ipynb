```python
'''
[Selenium 활용 예제]

연습 사이트
https://quotes.toscrape.com
'''
from selenium import webdriver
from selenium.webdriver.common.by import By

# ====================================
# 0) 브라우저 실행 + 페이지 접속
# ====================================
driver = webdriver.Chrome()
driver.get('https://quotes.toscrape.com')


# ====================================
# 1) 한 페이지에서 여러 명언 수집하기
# ====================================

print("===== 1페이지 수집 시작 =====")
# 수집한 데이터를 저장할 리스트
results = []

quote_blocks = driver.find_elements(By.CLASS_NAME, 'quote')

for block in quote_blocks:
    # 각 불록 안에서 필요한 데이터 찾기
    quote_text = block.find_element(By.CLASS_NAME, 'text').text
    author = block.find_element(By.CLASS_NAME, 'author').text

    # 태그는 여러 개 일 수 있으므로 find_elements를 사용
    tag_elements = block.find_elements(By.CLASS_NAME, 'tag')

    tags = [t.text for t in tag_elements]
    # 리스트 컴프리헨션을 사용하여 각 tag 요소의 텍스트만 추출
    #  - 컴프리헨션 : 반복문을 이용해 새로운 컬렉션을 만드는 축약 문법
    #  - [표현식 for 변수 in 반복가능객체]
    # for t in tag_elements:
    #     tags.append(t.text)

    results.append({
        'page': 1,
        'quote': quote_text,
        'author': author,
        'tags': tags
    })

    print(f"{author} - {quote_text} / tag:{tags}")

print('현재까지 수집 개수 :', len(results))

# 브라우저 종료
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

# ====================================
# 0) 브라우저 실행 + 페이지 접속
# ====================================
driver = webdriver.Chrome()
driver.get('https://quotes.toscrape.com')


# ====================================
# 2) 다음 페이지 이동 + 반복 수집
# ====================================

# 최대 몇 페이지까지 수집할지 제한하는 상수
# 무한 크롤링을 방지하기 위한 안전 장치
MAX_PAGES = 3

# 현재 페이지 번호를 추적하기 위한 변수
current_page = 1

results = []

while current_page <= MAX_PAGES:
    # 현재 수집 중인 페이지 번호 출력
    print(f"===== {current_page}페이지 수집 중 =====")

    # 새 페이지에서 다시 quote 블록 수집
    quote_blocks = driver.find_elements(By.CLASS_NAME, 'quote')

    for block in quote_blocks:
        quote_text = block.find_element(By.CLASS_NAME, 'text').text
        author = block.find_element(By.CLASS_NAME, 'author').text
        tag_elements = block.find_elements(By.CLASS_NAME, 'tag')
        tags = [t.text for t in tag_elements]

        results.append({
            'page': current_page,
            'quote': quote_text,
            'author': author,
            'tags': tags
        })

    # Next 버튼 찾기
    # "li.next a"는 class='next'를 가진 li안의 a태그를 의미
    # CSS_SELECTOR : CSS문법을 이용해서 HTML 요소를 찾는 방법
    # Selenium에서 가장 많이 쓰이는 방식 중 하나
    next_buttons = driver.find_elements(By.CSS_SELECTOR, 'li.next a')

    # Next 버튼이 없으면 종료
    if not next_buttons:
        print('다음 페이지가 없어 종료합니다.')
        break
    
    # 다음 페이지로 이동
    next_buttons[0].click()

    # 페이지 번호 1 증가
    current_page += 1

print('최종 수집 개수 :', len(results))
```

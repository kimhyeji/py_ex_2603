```python
# 어노테이션 : 어떤 타입인지 미리 적어두는 표시

# - 변수
age: int = 30
print(age)

# - 어노테이션은 강제 규칙이 아니다.
age : int = "안녕"
print(age)

# - 함수
# a는 int
# b는 int
# 결과도 int
def add(a:int, b:int) -> int:
    return a + b

# add(10, 20)
add('안녕', '반가워')

# - 리스트
nums: list[int] = [1, 2, 3]
print(nums)

# - 딕셔너리
person: dict[str, int] = {
    "age" : 30,
    "height" : 180
}

print(person)
```

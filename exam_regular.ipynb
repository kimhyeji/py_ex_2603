```python
# 파이썬에서 정규표현식을 사용하려면 re모듈을 불러와야 함
import re

text = '안녕하세요. 저는 15살 입니다, 저는 안경이 15개 있습니다.'

# r은 row string \(특수문자)
# \d는 숫자
# +는 1개 이상
pattern = r"\d+"

# findall(찾을규칙, 대상문자열)
# 문자열에서 패턴과 맞는 내용을 전부 찾아서 리스트로 반환
result = re.findall(pattern, text)
print(result)
```

```python
# 패턴 찾기
sample = 'abc123xyz456'

print(re.findall(r'\d+', sample))
print(re.findall(r'[a-z]+', sample))
```

```python
# 특정 형식 찾기

data = 'test123@gmail.com'

# \w : 문자 또는 숫자
# + : 1개 이상
# @ : 말 그대로 @
# w. : 점(특수 문자이기 때문에 \를 붙인다.)
# 문자/숫자 + @ + 문자/숫자 + . + 문자/숫자
email_pattern = r'\w+@\w+\.\w+'

emails = re.findall(email_pattern, data)
print(emails)
```

```python
# 문자열 치환

test = "가격은 1200원 입니다."

# sub(찾을패턴, 바꿀문자, 대상문자열)
# 문자열에서 특정 패턴을 찾아서 다른 문자로 바꾸는 함수
new_text = re.sub(r'\d+', '', test)
print(new_text)
```

```python
# 정규식 vs 일반 문자열 처리

test = '몸무게: 70kg, 키: 170cm'

test.split(',')
nums = re.findall(r'\d+', test)
print(nums)
```

```python
# 클래스 속성

class Person:
    folder = [] # 클래스 속성(공공제) -> 
    # 클래스에 속한다. 모든 인스턴스에서 접근이 가능하다.

    def add_files(self, file):
        Person.folder.append(file) # 클래스 이름으로 클래스 속성에 접근할 수 있다.

jin = Person()
jin.add_files('txt')

paul = Person()
paul.add_files('img')

print(jin.folder)
print(paul.folder)
```

```python
# 인스턴스 속성

class Person:
    def __init__(self):
        self.folder = [] # 인스턴스 속성

    def add_files(self, file):
        self.folder.append(file)   # 클래스 이름으로 클래스 속성에 접근할 수 있다.

jin = Person()
jin.add_files('txt')

paul = Person()
paul.add_files('img')

print(jin.folder)
print(paul.folder)
```

```python
# 비공개 클래스 속성

class Person:
    __weight = 78

    def print_weight(self):
        print(Person.__weight)

jin = Person()
jin.print_weight()

print(jin.__weight)
```

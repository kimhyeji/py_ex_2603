```python
# 클래스 -> 객체를 표현하기 위한 문법의 일종 , 프로그래밍으로 객체를 만들 때 사용하는 것

# 전사, 마법, 궁수, 자동차, 비행기, 사람, 고양이, 강아지 -> object
# -> 현실, 가상, 개념

# 속성(명사적 특징) - attribute
# 능력/기능(동사적 특징) - mehtod
# 전사 클래스 - 체력, 공격력, 무기 -> 속성
#            - 공격하다, 방어하다 -> 메서드
```

```python
# 클래스 -> 설계도
class Person:
    def Hello(self):
        print("hi")

jin = Person() # jin은 Person의 인스턴스(객체)
jin.Hello() # 인스턴스 메서드
```

```python
class Person:
    def Hello(self):
        print("Hello")
    
    def hi(self):
        self.Hello()

jin = Person()
jin.hi()
```
